```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset('iris')
data.head()
```

```python
plt.plot(data['sepal_width'])
```

```python
plt.plot(data['sepal_length'])
```

```python
plt.plot(data['sepal_width'],color='green', label="sepal width", linestyle='--')
plt.plot(data['sepal_length'],color='black', label="sepal length",marker='*')
plt.xlabel('Indexes',fontsize=10,color='red')
plt.ylabel('Petal length',fontsize=10,color='red')
plt.title('Plots',fontsize=20)
```

```python
plt.bar(data['species'],data['petal_length'])
plt.bar(data['species'],data['petal_width'])
plt.plot()
```
